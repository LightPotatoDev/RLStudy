# 2.1 마르코프 결정 과정

**에이전트의 행동에 따라 상태가 변함** (MDP의 필요 이유)<br>

특정 시간에 에이전트가 행동을 취하고, 그 결과 새로운 상태로 전이<br>
이때의 시간 단위를 **타임 스텝**이라고 함

image.png

상태 $S_t$에서 에이전트가 행동 $A_t$를 수행 -> 보상 $R_t$ 획득 & 상태 $S_{t+1}$로 전이

# 2.2 환경과 에이전트를 수식으로

[상태 전이의 수식 표현] <br>
결정적(deterministic) - 다음 상태 $s'$은 현재 상태 $s$와 행동 $a$에 의해 단 하나로 결정<br>
$s' = f(s,a)$ (상태 전이 함수)<br>
확률적(stochastic) - 다음 상태는 여러 가지 중 하나가 될 수 있음<br>
$p(s'|s,a)$ - 현재 상태 s에서 행동 a를 취하면 s'으로 넘어갈 확률 (상태 전이 확률)<br><br>
마르코프 성질 - 현재의 정보만 고려 (과거의 상태와 행동은 고려하지 않음)

[보상 함수]<br>
$r(s,a,s')$ - 상태 s에서 행동 a를 수행해 다음 상태가 s'이 되었을 때 얻는 보상<br>

[정책]<br>
현재 상태만으로 에이전트가 행동을 결정할 수 있음<br>
결정적 정책 - 상태 s에서 반드시 행동 a를 함 $a = \mu(s)$<br>
확률적 정책 - 상태 s에서 확률적으로 행동 a를 함 / 확률: $\pi(a|s)$

# 2.3 MDP의 목표

일회성 과제 (episodic task) - 끝이 있는 문제 (바둑 등)
 - 에피소드: 시작부터 끝까지의 일련의 시도

지속적 과제 (continuous task) - 끝이 없는 문제 (재고 관리 등)


수익 (return) - 에이전트는 **최적 정책**을 통해 이를 최대화시켜야 함

$G_t = R_t + \gamma R_{t+1} + \gamma^2 R_{t+2} + ...$

$\gamma$ - 할인율(0-1 사이의 실수)
 - 지속적 과제에서 수익을 무한대로 만들지 않기 위해 도입
 - 가까운 미래의 보상을 더 중요하게 보이도록 함

[?] 할인율이 더 먼 미래의 보상을 더 중요하게 보이도록 하는 수익 계산 방법이 있을까?

상태 가치 함수 - 수익의 기댓값을 행동 지표로 삼음 (행동, 상태 전이가 확률적으로 일어나기 때문)

$v_{\pi}(s) = E[G_t|S_t = s,\pi]$

상태 $S_t$가 s이고, 정책이 $\pi$일 때 에이전트가 얻을 수 있는 기대 수익
<br><br>
모든 상태 s에서 $v_{\pi'}(s) >= v_{\pi}(s)$ 가 성립할 시 ${\pi'}$이 더 나은 정책이고 판단함.<br>
가장 나은 정책을 ${\pi}_*$으로 표현, 이를 결정적 정책이라 함<br>
결정적 정책에서는 각 상태에서의 행동이 유일하게 결정됨<br>

$\alpha = \mu_* (s)$
상태 s에서 행동 a를 출력

최적 상태 가치 함수 - 최적 정책의 상태 가치 함수 $v_*$
